## Outliers: Checking

```python
def outliers(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    outlier_mask = pd.DataFrame(np.zeros(df.shape),columns=df.columns)
    outlier_mask.index = df.index
    for col in df.columns:
        outlier_mask[col] = (df[col] < (Q1[col] - 1.5 * IQR[col])) | (df[col] > (Q3[col] + 1.5 * IQR[col]))
    return df[outlier_mask.any(axis=1)]
```

```python
outliers=outliers(data)
#outliers
```

- We've identified a total of <b>30,6% of data</b> that is <b>considered outliers</b>. This is <b>almost a third</b> of the dataset, so it <b>wouldn't</b> be smart to <b>drop them all</b>.

## SCALING: Yeo-Johnson Transformation

<b>Yeo-Johnson transformation</b> is a <b>variant of the Box-Cox</b> transformation which <b>allows negative values</b> by using the absolute value of the input and for negative input multiply it by -1.

```python
def yeo_johnson(x, lambda_):
    if x > 0:
        y = (x**lambda_ - 1) / lambda_
    else:
        y = -1 * (((-1*x)**lambda_ - 1) / lambda_)
    return y
```

```python
data_yj=pd.DataFrame()
for column in data.columns:
    data_yj[column] = data[column].apply(lambda x: yeo_johnson(x, lambda_ = 0.5))
```

## SCALING: Power Transformation

```python
def power_transform(x, p):
    return x**p
```

```python
data_pt = pd.DataFrame()
for column in data.columns:
    data_pt[column] = data[column].apply(lambda x: power_transform(x, p=2))
```

## SCALING: Z-Score Transformation

```python
def z_score(df):
    mean = cont.mean()
    std = cont.std()
    df_zscore = (df - mean) / std
    return df_zscore
```

```python
data_z=z_score(data)
```

### Comparing the Different Transformations

```python
for column in data.columns:
    fig, axes = plt.subplots(2, 2, figsize=(10, 7))
    fig.suptitle(column)
    sns.distplot(data[column], ax=axes[0][0], kde=True, color = 'gray').set_xlabel( 'No Transformation' , size = 12 )
    sns.distplot(data_yj[column], ax=axes[0][1], kde=True, color = 'gold').set_xlabel( 'Yeo-Johnson Transformation' , size = 12 )
    sns.distplot(data_pt[column], ax=axes[1][0], kde=True, color = 'orangered').set_xlabel( 'Power Transformation' , size = 12 )
    sns.distplot(data_z[column], ax=axes[1][1], kde=True, color = 'royalblue').set_xlabel( 'Z-score Transformation' , size = 12 )
    plt.show()
```

# <CENTER> MODELING AREA

## Modeling: Splitting the Data

```python
X = data.drop('price', axis=1)
y = data['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## Modeling: Calling the Models

```python
def lr():
    model = LinearRegression()
    model.fit(X, y)
    return model

def knr():
    model = KNeighborsRegressor(n_neighbors=5)
    model.fit(X, y)
    return model

def dtr():
    model = DecisionTreeRegressor()
    model.fit(X, y)
    return model

def rfr():
    model = RandomForestRegressor()
    model.fit(X, y)
    return model

def gbr():
    model = GradientBoostingRegressor()
    model.fit(X, y)
    return model

def xgbr():
    model = XGBRegressor()
    model.fit(X, y)
    return model

def lgbmr():
    model = LGBMRegressor()
    model.fit(X, y)
    return model
```

```python
models = [lr(),knr(),dtr(),rfr(),gbr(),xgbr(),lgbmr()]
```

## Modeling: Training the Models

```python
def train_test_models(models, X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    results = []
    for model in models:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        results.append((model, mse, r2)) 
    return results
```

## Modeling: Results

```python
results = train_test_models(models, X, y)
```

## Modeling: Compare the Models

```python
model_names = [model[0].__class__.__name__ for model in results]
mse_scores = [model[1] for model in results]
r2_scores = [model[2] for model in results]

x = range(len(model_names))

fig, ax = plt.subplots()
ax.grid(True)
ax.bar(x, mse_scores,color='royalblue')
fig.set_size_inches(12, 8)
plt.xticks(x, model_names,rotation=45)
plt.title('MSE Scores')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(x, r2_scores,color='orangered')
ax.grid(True)
fig.set_size_inches(12, 8)
plt.xticks(x, model_names,rotation=45)
plt.title('R-squared Scores')
plt.show()
```
